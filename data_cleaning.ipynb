```python
import pandas as pd
```

```python
est_cases = pd.read_csv('MALARIA_EST_CASES.csv', skiprows=1)
est_cases = est_cases.rename(columns={
    'Countries, territories and areas': 'region',
    'Year': 'year',
    'No. of cases': 'cases'
})
est_cases = est_cases.melt(id_vars='region', var_name='year', value_name='cases')
est_cases.head()
```

```python
est_deaths = pd.read_csv('MALARIA_EST_DEATHS.csv', skiprows=1)
est_deaths = est_deaths.rename(columns={
    'Countries, territories and areas': 'region',
    'Year': 'year',
    'No. of deaths': 'deaths'
})
est_deaths = est_deaths.melt(id_vars='region', var_name='year', value_name='deaths')
est_deaths.head()
```

```python
est_no = pd.merge(est_cases, est_deaths)
est_no.head()
```

```python
for col in ['cases', 'deaths']:
    est_no[col] = est_no[col].str.replace(' ', '')
    est_no[col+'_median'] = est_no[col].str.extract('([0-9.]+)')
    est_no[col+'_min'] = est_no[col].str.extract('[0-9.]+\[([0-9.]+)')
    est_no[col+'_max'] = est_no[col].str.extract('[0-9.]+\[[0-9.]+\-([0-9.]+)')
    
est_no.head()
```
